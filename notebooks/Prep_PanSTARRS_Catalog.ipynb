```python
from astropy.table import Table
from astropy.io import ascii
import numpy as np
```

```python
in_fn = '/Users/kirstencasey/Downloads/NGC672dwD_inputimages/panstarrs-27.5-27.5.csv'
out_fn = '/Users/kirstencasey/Downloads/NGC672dwD_inputimages/panstarrs-27.5-27.5_withBESSEL.dat'

t = Table.read('/Users/kirstencasey/projects/LBCreduce/lbcred/utils/panstarrs_bandpass_transformations.dat',format='ascii')
panstarrs = Table.read(in_fn)
panstarrs.rename_column('raMean', 'ra')
panstarrs.rename_column('decMean', 'dec')
ascii.write(panstarrs,in_fn,overwrite=True,format='csv')
```

```python
panstarrs_bessel = panstarrs.copy()
panstarrs_bessel['R-BESSEL_linear'] = np.zeros(len(panstarrs))
panstarrs_bessel['B-BESSEL_linear'] = np.zeros(len(panstarrs))
panstarrs_bessel['R-BESSEL_quad'] = np.zeros(len(panstarrs))
panstarrs_bessel['B-BESSEL_quad'] = np.zeros(len(panstarrs))
```

```python
i = 0
for obj in panstarrs_bessel:
    x = obj['gMeanPSFMag'] - obj['rMeanPSFMag']
    
    R_lin = t[1]['B0'] + (t[1]['B1'] * x) + obj['rMeanPSFMag']
    R_quad = t[1]['A0'] + (t[1]['A1'] * x) + (t[1]['A2'] * x**2) + obj['rMeanPSFMag']
    B_lin = t[0]['B0'] + (t[0]['B1'] * x) + obj['gMeanPSFMag']
    B_quad = t[0]['A0'] + (t[0]['A1'] * x) + (t[0]['A2'] * x**2) + obj['gMeanPSFMag']

    panstarrs_bessel[i]['R-BESSEL_linear'] = R_lin 
    panstarrs_bessel[i]['R-BESSEL_quad'] = R_quad
    panstarrs_bessel[i]['B-BESSEL_linear'] = B_lin 
    panstarrs_bessel[i]['B-BESSEL_quad'] = B_quad
    i+=1

panstarrs_bessel
```

```python
ascii.write(panstarrs_bessel,out_fn)
```
